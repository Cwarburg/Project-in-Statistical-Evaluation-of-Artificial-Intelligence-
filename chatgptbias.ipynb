```python


# Define a function to calculate and print Chi-Square statistics for gender and ethnicity
def chi_square_analysis(df, label):
    print(f'\n{label} Gender')
    
    # Gender analysis
    gender_counts = np.array([df['Male'].sum(), df['Female'].sum()])
    total_gender_counts = gender_counts.sum()
    expected_gender_counts = np.array([total_gender_counts / 2, total_gender_counts / 2])
    
    chi2_stat_gender, p_value_gender = chisquare(gender_counts, expected_gender_counts)
    print(f'Chi-square statistic: {chi2_stat_gender}\nP-value: {p_value_gender}')
    
    # Ethnicity analysis
    print(f'\n{label} Ethnicity')
    ethnicity_counts = np.array([df['White'].sum(), df['Hispanic'].sum(), df['Black'].sum(), df['Asian'].sum()])
    total_ethnicity_counts = ethnicity_counts.sum()
    expected_ethnicity_counts = np.array([total_ethnicity_counts / 4, total_ethnicity_counts / 4, total_ethnicity_counts / 4, total_ethnicity_counts / 4])
    
    chi2_stat_ethnicity, p_value_ethnicity = chisquare(ethnicity_counts, expected_ethnicity_counts)
    print(f'Chi-square statistic: {chi2_stat_ethnicity}\nP-value: {p_value_ethnicity}')

# List of DataFrames and their corresponding labels
dfs = [
    (df_doctor, 'Doctor'),
    (df_model, 'Model'),
    (df_nurse, 'Nurse'),
    (df_engineer, 'Engineer')
    # Assuming df_engineer and df_nurse are defined similarly
    # (df_engineer, 'Engineer'),
    # (df_nurse, 'Nurse')
]

# Iterate over each DataFrame and perform the Chi-Square analysis
for df, label in dfs:
    chi_square_analysis(df, label)

print(df_doctor['Male'].sum())

```

```python
def plot_gender_distribution_with_ci(df, ax, title):
    total_male = df['Male'].sum()
    total_female = df['Female'].sum()
    total = total_male + total_female

    male_percent = total_male / total
    female_percent = total_female / total

    confidence_level = 0.95
    n = total

    z = stats.norm.ppf(1 - (1 - confidence_level) / 2)

    male_se = np.sqrt(male_percent * (1 - male_percent) / n)
    female_se = np.sqrt(female_percent * (1 - female_percent) / n)

    male_ci = z * male_se
    female_ci = z * female_se

    male_lower = (male_percent - male_ci) * 100
    male_upper = (male_percent + male_ci) * 100
    female_lower = (female_percent - female_ci) * 100
    female_upper = (female_percent + female_ci) * 100

    categories = ['Male', 'Female']
    percentages = [male_percent * 100, female_percent * 100]
    errors = [male_ci * 100, female_ci * 100]

    bars = ax.bar(categories, percentages, yerr=errors, capsize=10, color=['green', 'blue'])
    ax.set_ylabel('Percentage')
    ax.set_title(title)
    
    # Add text for confidence intervals
    for bar, lower, upper in zip(bars, [male_lower, female_lower], [male_upper, female_upper]):
        height = bar.get_height()
        ax.text(bar.get_x() + bar.get_width() / 2, height + 1, f'[{lower:.2f}, {upper:.2f}]', ha='center', va='bottom')

# Assuming df_engineer, df_model, df_nurse, and df_doctor are defined

fig, axs = plt.subplots(2, 2, figsize=(15, 10))

plot_gender_distribution_with_ci(df_engineer, axs[0, 0], 'Engineer Gender proportion ')
plot_gender_distribution_with_ci(df_model, axs[0, 1], 'Model Gender proportion')
plot_gender_distribution_with_ci(df_nurse, axs[1, 0], 'Nurse Gender proportion')
plot_gender_distribution_with_ci(df_doctor, axs[1, 1], 'Doctor Gender proportion')

plt.tight_layout()
plt.show()
```
